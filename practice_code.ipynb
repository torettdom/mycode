```python
#Plot the total sum of Transaction_Amount per Account_Type using a bar plot.

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Account_Type': ['Savings', 'Checking', 'Savings', 'Credit', 'Checking'],
    'Transaction_Amount': [200, 150, 300, 400, 100]
}
df = pd.DataFrame(data)
grouped = df.groupby('Account_Type')['Transaction_Amount'].sum()

plt.figure(figsize=(8, 5))
grouped.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Create a pie chart to show the percentage of transactions per Branch.
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Branch': ['North', 'South', 'East', 'West', 'North', 'South'],
    'Transaction_Amount': [200, 150, 300, 400, 100, 250]
}
df = pd.DataFrame(data)
grouped = df.groupby('Branch')['Transaction_Amount'].sum()

plt.figure(figsize=(8, 8))
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')
plt.show()
```
