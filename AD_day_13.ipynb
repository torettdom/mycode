```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris_data = sns.load_dataset('iris')
df = sns.load_dataset('taxis')
```

```python
sns.scatterplot(x=iris_data['sepal_length'], y=iris_data['sepal_width'])
plt.show()
```

```python
sns.scatterplot(x=iris_data['species'],y=iris_data['sepal_width'],palette=['r','g','y'])
plt.show()
```

```python
sns.stripplot(x=iris_data['species'],y=iris_data['sepal_width'],palette=['r','g','y'])
plt.show()
```

```python
sns.stripplot(x=iris_data['species'],y=iris_data['sepal_width'],palette=['r','g'])
plt.show()
```

```python
sns.stripplot(x=df['pickup_borough'],y=df['tip'],palette=['r','g'])
plt.show()
```

```python
sns.lineplot(x=df['tip'],y=df['fare'])
plt.show()
```

```python
sns.lineplot(x=df['total'],y=df['fare'],hue=df['payment'])
plt.show()
```

```python
sns.lineplot(x=df['total'],y=df['fare'],hue=df['payment'],palette=['red','orange'])
plt.show()
```

```python
plt.figure(figsize=(10,5))#we can change the size using figsize
sns.countplot(x=df['payment'],hue=df['color'],palette=['green','orange'])
```

```python
plt.figure(figsize=(10,5))#we can change the size using figsize
sns.countplot(x=df['payment'],hue=df['color'],palette=['green','orange'])
plt.xlabel('Payment',fontsize=28,color='blue')
plt.ylabel('Count')
plt.title('Payment Count')
```
